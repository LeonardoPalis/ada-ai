```python
from fastai.tabular.all import *
from numpy.random import normal, seed, uniform
from ipywidgets import interact, FloatSlider
from functools import partial
```

```python
np.random.seed(42)
def noise(x, scale): return normal(scale=scale, size=x.shape)
def add_noise(x, mult, add): return x * (1+noise(x,mult) + noise(x, add))
def quad(a,b,c,x): return a*x**2 + b*x + c
def mk_quad(a,b,c): return partial(quad, a,b,c)
def quad_mse(params):
  f = mk_quad(*params)
  return mse(f(x), y)

def plot_function(func, xlim=(-10, 10), ylim=None, num_points=100, title="Function Plot"):
    """
    Plots a given mathematical function.

    Parameters:
        func (function): The mathematical function to plot.
        xlim (tuple): The range of x values (default is (-10, 10)).
        ylim (tuple): The range of y values (default is None, which auto-scales).
        num_points (int): Number of points to calculate for the plot (default is 100).
    """
    # Generate x values
    x = np.linspace(xlim[0], xlim[1], num_points)
    # Calculate y values
    y = func(x)

    # Create the plot
    plt.plot(x, y, label="Function")
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    if ylim:
        plt.ylim(ylim)
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')

# Definindo os termos extraídos do documento e da consulta
doc_terms = {"carro": (0.2, 0.7), "estacionado": (0.2, 0.3)}
query_terms = {"automóvel": (0.8, 0.7), "estacionado": (0.8, 0.3)}

# Exibe as caixas para os termos do documento
for term, pos in doc_terms.items():
    ax.text(pos[0], pos[1], f"Doc: {term}", fontsize=12, ha='center',
            bbox=dict(facecolor='lightblue', alpha=0.5, edgecolor='black'))

# Exibe as caixas para os termos da consulta
for term, pos in query_terms.items():
    ax.text(pos[0], pos[1], f"Query: {term}", fontsize=12, ha='center',
            bbox=dict(facecolor='lightgreen', alpha=0.5, edgecolor='black'))

# Conexão para o termo que realmente coincide ("estacionado")
ax.annotate("", xy=(query_terms["estacionado"][0] - 0.05, query_terms["estacionado"][1]),
            xytext=(doc_terms["estacionado"][0] + 0.05, doc_terms["estacionado"][1]),
            arrowprops=dict(arrowstyle="->", lw=1.5, color='black'))

# Tenta conectar termos semanticamente similares mas que não coincidem textualmente:
# "carro" (documento) e "automóvel" (consulta) são semanticamente equivalentes,
# mas na indexação invertida não há ligação entre eles.
ax.annotate("Falha semântica",
            xy=((doc_terms["carro"][0] + query_terms["automóvel"][0]) / 2,
                (doc_terms["carro"][1] + query_terms["automóvel"][1]) / 2),
            fontsize=12, color='red', ha='center')

ax.annotate("", xy=(query_terms["automóvel"][0] - 0.05, query_terms["automóvel"][1]),
            xytext=(doc_terms["carro"][0] + 0.05, doc_terms["carro"][1]),
            arrowprops=dict(arrowstyle="->", lw=1.5, linestyle='dashed', color='red'))

plt.title("Limitação da Indexação Invertida: Sinônimos Não Conectados", fontsize=14)
plt.show()
```
